```python
import numpy as np
import matplotlib.pyplot as plt
from DBSCAN import *
```

```python
csv = np.genfromtxt("e7.csv", delimiter=',', skip_header=1, dtype=float)
np.random.shuffle(csv)
t_data = csv[:,:]
```

```python
lr = DBSCAN(t_data)
k=4
dist = lr.find_k_distance(k)
plt.figure(figsize=(8, 6))
plt.plot(dist, marker='o', linestyle='-', color='b')
plt.title(f'K-Distance Graph (k={k})')
plt.xlabel('Points')
plt.ylabel(f'{k}-th Nearest Neighbor Distance')
plt.show()
```

```python
lr.train(0.1,4)
```

```python
plt.figure(figsize=(8, 6))
colors = ['blue', 'yellow','green','red']
for cluster_id,cluster in enumerate(lr.clusters):
    plt.scatter(cluster[:,0], cluster[:, 1], color=colors[cluster_id], alpha=0.5)

plt.title('2D Scatter Plot of Data with Labels (Features 0 and 1)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 3')
plt.show()
```
